```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

# set seaborn style
sns.set_theme()

def plot_data(d):
    time = [x["time"] for x in d]
    val  = [x["value"] for x in d]

    plt.plot(time, val, "-o", markersize=2)
    plt.ylim(0, None)

def plot_data_year(d):
    time = np.array([x["time"] for x in d])
    val  = np.array([x["value"] for x in d])

    t = time[:-1:4]
    v = val[:-1:4] + val[1:-1:4] + val[2:-1:4] + val[3:-1:4]

    plt.plot(t, v)
    plt.ylim(0, None)
```
